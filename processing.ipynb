```python
import pandas as pd
import torch
import nltk
from eunjeon import Mecab
import numpy as np
from nltk import FreqDist
from torchtext.legacy import data,datasets
import os
from torchtext.legacy.data import  TabularDataset, BucketIterator, Iterator
df=pd.read_excel('comunications.xlsx')
tokenizer=Mecab()
question,answer=df['Q'],df['A']

PAD, BOS, EOS = 1, 2, 3

class DataLoader():

    def __init__(self,
                 train_fn=None,
                 valid_fn=None,
                 exts=None,
                 batch_size=64,
                 device='cpu',
                 max_vocab=99999999,
                 max_length=255,
                 fix_length=None,
                 use_bos=True,
                 use_eos=True,
                 shuffle=True,
                 dsl=False
                 ):

        super(DataLoader, self).__init__()

        self.src = data.Field(
            sequential=True,
            use_vocab=True,
            batch_first=True,
            include_lengths=True,
            fix_length=fix_length,
            init_token='<BOS>' if dsl else None,
            eos_token='<EOS>' if dsl else None,
        )

        self.tgt = data.Field(
            sequential=True,
            use_vocab=True,
            batch_first=True,
            include_lengths=True,
            fix_length=fix_length,
            init_token='<BOS>' if use_bos else None,
            eos_token='<EOS>' if use_eos else None,
        )

        if train_fn is not None and valid_fn is not None and exts is not None:
            train = TranslationDataset(
                path=train_fn,
                exts=exts,
                fields=[('src', self.src), ('tgt', self.tgt)],
                max_length=max_length
            )
            valid = TranslationDataset(
                path=valid_fn,
                exts=exts,
                fields=[('src', self.src), ('tgt', self.tgt)],
                max_length=max_length,
            )

            self.train_iter = data.BucketIterator(
                train,
                batch_size=batch_size,
                device='cuda:%d' % device if device >= 0 else 'cpu',
                shuffle=shuffle,
                sort_key=lambda x: len(x.tgt) + (max_length * len(x.src)),
                sort_within_batch=True,
            )
            self.valid_iter = data.BucketIterator(
                valid,
                batch_size=batch_size,
                device='cuda:%d' % device if device >= 0 else 'cpu',
                shuffle=False,
                sort_key=lambda x: len(x.tgt) + (max_length * len(x.src)),
                sort_within_batch=True,
            )

            self.src.build_vocab(train, max_size=max_vocab)
            self.tgt.build_vocab(train, max_size=max_vocab)

    def load_vocab(self, src_vocab, tgt_vocab):
        self.src.vocab = src_vocab
        self.tgt.vocab = tgt_vocab


class TranslationDataset(data.Dataset):
    """Defines a dataset for machine translation."""

    @staticmethod
    def sort_key(ex):
        return data.interleave_keys(len(ex.src), len(ex.trg))

    def __init__(self, path, exts, fields, max_length=None, **kwargs):
        """Create a TranslationDataset given paths and fields.
        Arguments:
            path: Common prefix of paths to the data files for both languages.
            exts: A tuple containing the extension to path for each language.
            fields: A tuple containing the fields that will be used for data
                in each language.
            Remaining keyword arguments: Passed to the constructor of
                data.Dataset.
        """
        if not isinstance(fields[0], (tuple, list)):
            fields = [('src', fields[0]), ('trg', fields[1])]

        if not path.endswith('.'):
            path += '.'

        src_path, trg_path = tuple(os.path.expanduser(path + x) for x in exts)

        examples = []
        with open(src_path, encoding='utf-8') as src_file, open(trg_path, encoding='utf-8') as trg_file:
            for src_line, trg_line in zip(src_file, trg_file):
                src_line, trg_line = src_line.strip(), trg_line.strip()
                if max_length and max_length < max(len(src_line.split()), len(trg_line.split())):
                    continue
                if src_line != '' and trg_line != '':
                    examples += [data.Example.fromlist([src_line, trg_line], fields)]

        super().__init__(examples, fields, **kwargs)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from eunjeon import Mecab
import torch.nn as nn
from torchtext.legacy import data,datasets
from torchtext.legacy.data import  TabularDataset, BucketIterator, Iterator
import pandas as pd
# 문자 단위 RNN 구현 예제
input_str = 'apple'
label_str = 'pple!'
char_vocab = sorted(list(set(input_str+label_str)))
vocab_size = len(char_vocab)
print ('문자 집합의 크기 : {}'.format(vocab_size))
input_size=vocab_size
# 단어 사전 형성
hidden_size=5
output_size=5
learning_rate=0.1
char_toindex=dict((c,i) for i, c in enumerate(char_vocab))
print(char_toindex)
index_to_char={}
for key, value in char_toindex.items():
    index_to_char[value] = key
print(index_to_char)
x_data = [char_toindex[c] for c in input_str]
y_data = [char_toindex[c] for c in label_str]
x_data=[x_data]
y_data=[y_data]
x_onehot=[np.eye(vocab_size)[x]for x in x_data]
# 텐서로 바꿔주기. 
x=torch.FloatTensor(x_onehot)
y=torch.LongTensor(y_data)
print(x.shape)
print(y.shape)
# 모델 구현
class Net(torch.nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(Net,self).__init__()
        self.rnn=torch.nn.RNN(input_size, hidden_size, batch_first=True)
        self.fc=torch.nn.Linear(hidden_size, output_size, bias=True)
    def forward(self, x):
        x,_status=self.rnn(x)
        x=self.fc(x)
        return x
net=Net(input_size, hidden_size, output_size)
outputs=net(x)
print(outputs.shape)
# 펼치기
print(outputs.view(-1, input_size).shape)
print(y.shape)
print(y.view(-1).shape)
# 손실함수 정의
criterion=torch.nn.CrossEntropyLoss()
optimizer=optim.Adam(net.parameters(),learning_rate)
# 반복 학습을 통해서 학습
for i in range(100):
    optimizer.zero_grad()
    outputs=net(x)
    # 배치 차원 제거
    loss=criterion(outputs.view(-1,input_size),y.view(-1))
    loss.backward() # 기울기 계산
    optimizer.step() # 파라미터들 업데이트
    result=outputs.data.numpy().argmax(axis=2) # 제일 높은 인덱스 return
    result_str = ''.join([index_to_char[c] for c in np.squeeze(result)]) # 단어 사전 통해서 단어로 복원. 
    print(i, "loss(crossentropy) ", loss.item(), "prediction word ", result, "true Y: ", y_data, "prediction str: ", result_str)
```

```python
import torch
import os
from eunjeon import Mecab
import torch.nn as nn
import torch.nn.functional as f
from torchtext.legacy import data,datasets
from torchtext.legacy.data import  TabularDataset, BucketIterator, Iterator
import random
import pandas as pd
import torch.optim as optim
SOS_token=0
EOS_token=1
df=pd.read_excel('comunications.xlsx')
df['Q'] = df['Q'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
df['A'] = df['A'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
train_df=df[:1000]
test_df=df[1000:]
train_df.to_csv("train_data.csv",index=False)
test_df.to_csv("test_data.csv",index=False)
torch.manual_seed(0)
device=torch.device('cuda' if torch.cuda.is_available() else'cpu')

class Vocab:
    def __init__(self):
        self.vocab2index={"<SOS>":SOS_token,"<EOS>":EOS_token}
        self.index2vocab={SOS_token:"SOS",EOS_token:"<EOS>"}
        self.vocab_count={}
        self.n_vocab=len(self.vocab2index)
    def add_vocab(self,sentence):
        for word in sentence.split(' '):
            if word not in self.vocab2index:
                self.vocab2index[word]=self.n_vocab
                self.vocab_count[word]=1
                self.index2vocab[self.n_vocab]=word
                self.n_vocab+=1
            else:
                self.vocab_count[word]+=1

class Encoder(nn.Module):
    def __init__(self,input_size, hidden_size):
        super(Encoder,self).__init__()
        self.hidden_size=hidden_size
        self.embedding=nn.Embedding(input_size, hidden_size)
        self.gru=nn.GRU(hidden_size,hidden_size)
    def forward(self,x,hidden):
        x=self.embedding(x).view(1,1,-1)
        x,hidden=self.gru(x,hidden)
        return x,hidden

class Decoder(nn.Module):
    def __init__(self,hidden_size,output_size):
        super(Decoder,self).__init__()
        self.hidden_size=hidden_size
        self.embedding=nn.Embedding(output_size,hidden_size)
        self.gru=nn.GRU(hidden_size, hidden_size, num_layers=1, batch_first=True)
        self.out=nn.Linear(hidden_size, output_size)
        self.softmax=nn.LogSoftmax(dim=1)
    def forward(self,x, hidden):
        x=self.embedding(x).view(1,1,-1)
        x,hidden=self.gru(x, hidden)
        x=self.softmax(self.out(x[0]))
        return x, hidden
# 소스 데이터와 타겟 데이터 설정. 
def preprocess(df):
    pairs=[]
    source_vocab=Vocab()
    target_vocab=Vocab()
    for i in df.index:
        source_vocab.add_vocab(df['Q'][i])
        target_vocab.add_vocab(df['A'][i])
        temp=[df['Q'][i],df['A'][i]]
        pairs.append(temp)
    
    return source_vocab, target_vocab,pairs

source_vocab, target_vocab, pairs=preprocess(df)

enc_hidden_size=30
dec_hidden_size=enc_hidden_size
enc=Encoder(source_vocab.n_vocab,enc_hidden_size)
dec=Decoder(dec_hidden_size,target_vocab.n_vocab)

# 텐서화. 
def tensorize(vocab,sentence):
    idx=[vocab.vocab2index[word] for word in sentence.split(' ')]
    idx.append(vocab.vocab2index['<EOS>'])
    return torch.Tensor(idx).long().to(device).view(-1,1)

# 중간점검
print(tensorize(source_vocab,'좋은 아침'))
print(tensorize(target_vocab,'좋은 아침이에요'))
train_source=[tensorize(source_vocab,pair[0]) for pair in pairs]
train_target=[tensorize(target_vocab,pair[1]) for pair in pairs]
loss_total=0
number_epoch=100
# 학습시키기. 
encoder_optimizer=optim.SGD(enc.parameters(),lr=0.01)
decoder_optimizer=optim.SGD(dec.parameters(),lr=0.01)
criterion=nn.NLLLoss()
for epoch in range(number_epoch):
    epoch_loss=0
    for i in range(len(train_source)):
        encoder_optimizer.zero_grad()
        decoder_optimizer.zero_grad()
        source_tensor=train_source[i]
        target_tensor=train_target[i]
        encoder_hidden=torch.zeros([1,1,enc.hidden_size]).to(device)
        source_length=source_tensor.size(0)
        target_length=target_tensor.size(0)
        loss=0
        for enc_input in range(source_length):
            _, encoder_hidden = enc(source_tensor[enc_input], encoder_hidden)
        decoder_input = torch.Tensor([[SOS_token]]).long().to(device)
        decoder_hidden = encoder_hidden 
        for count in range(target_length):
            decoder_output, decoder_hidden = dec(decoder_input, decoder_hidden)       
            loss += criterion(decoder_output, target_tensor[count])
            decoder_input = target_tensor[count]  
        
        loss.backward()
        encoder_optimizer.step()
        decoder_optimizer.step()
        epoch_loss += loss.item()/target_length
    if epoch % 100 == 0:
        print('--- epoch {}, total loss {} '.format(epoch,float(epoch_loss/15)))
# 평가하기
for pair in pairs:
    source_tensor=tensorize(source_vocab, pair[0])
    source_length=source_tensor.size()[0]
    encoder_hidden=torch.zeros([1,1,enc.hidden_size]).to(device)
    for ecount in range(source_length):
        _,encoder_hidden=enc(source_tensor[ecount],encoder_hidden)
    decoder_input = torch.Tensor([[SOS_token]], device=device).long()
    decoder_hidden = encoder_hidden
    decoded_words = []
    for dcount in range(20):
        decoder_output,decoder_hidden=dec(decoder_input,decoder_hidden)
        _, top_index=decoder_output.data.topk(1)
        if top_index.item()==EOS_token:
            decoded_words.append('<EOS>')
            break
        else:
            decoded_words.append(target_vocab.index2vocab[top_index.item()])
            
        decoder_input=top_index.squeeze().detach()
    predict_words=decoded_words
    predict_sentence=" ".join(predict_words)
    print("Q",pair[0])
    print("A",pair[1])
    print("Predict words",predict_sentence)
```

```python
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as f
import torch.optim as optim
device=torch.device("cuda" if torch.cuda.is_available() else "cpu")
SOS_token=0
EOS_token=1
df=pd.read_excel('comunications.xlsx')
df['Q'] = df['Q'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
df['A'] = df['A'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
train_df=df[:1000]
test_df=df[1000:]
train_df.to_csv("train_data.csv",index=False)
test_df.to_csv("test_data.csv",index=False)
torch.manual_seed(0)
MAX_LENGTH=10 # 어텐션 개념을 사용하려면 max_length를 설정해주어야 한다. 
device=torch.device('cuda' if torch.cuda.is_available() else'cpu')
class Vocab:
    def __init__(self):
        self.vocab2index={"<SOS>":SOS_token,"<EOS>":EOS_token}
        self.index2vocab={SOS_token:"SOS",EOS_token:"<EOS>"}
        self.vocab_count={}
        self.n_vocab=len(self.vocab2index)
    def add_vocab(self,sentence):
        for word in sentence.split(' '):
            if word not in self.vocab2index:
                self.vocab2index[word]=self.n_vocab
                self.vocab_count[word]=1
                self.index2vocab[self.n_vocab]=word
                self.n_vocab+=1
            else:
                self.vocab_count[word]+=1
class EncoderA(nn.Module):
    def __init__(self, input_size, hidden_size):
        super(EncoderA,self).__init__()
        self.hidden_size=hidden_size
        self.embedding=nn.Embedding(input_size, hidden_size)
        self.gru=nn.GRU(hidden_size, hidden_size)
    def forward(self, input, hidden):
        embedded=self.embedding(input).view(1,1,-1)
        output=embedded
        output, hidden=self.gru(output,hidden)
        return output, hidden
    def initHidden(self):
        return torch.zeros(1,1, self.hidden_size, device=device)
class Decoder(nn.Module):
    def __init__(self, hidden_size, output_size):
        super(Decoder,self).__init__()
        self.hidden_size=hidden_size
        self.embedding=nn.Embedding(output_size, hidden_size)
        self.gru=nn.GRU(hidden_size,hidden_size)
        self.out=nn.Linear(hidden_size, output_size)
        self.softmax=nn.LogSoftmax(dim=1)
    def forward(self, input, hidden):
        output=self.embedding(input).view(1,1,-1)
        output=f.relu(output)
        output, hidden=self.gru(output, hidden)
        output=self.softmax(self.out(output[0]))
        return output, hidden
    def initHidden(self):
        return torch.zeros(1,1, self.hidden_size, device=device)
# 어텐션 디코더 ( 그 전 디코더 보다 성능 좋게)
class AttnDecoder(nn.Module):
    def __init__(self, hidden_size, output_size, dropout_p=0.1,max_length=MAX_LENGTH):
        super(AttnDecoder, self).__init__()
        self.hidden_size=hidden_size
        self.output_size=output_size
        self.dropout_p=dropout_p
        self.max_length=max_length
        self.embedding=nn.Embedding(self.output_size, self.hidden_size)
        self.attn=nn.Linear(self.hidden_size *2, self.max_length)
        self.attn_combine=nn.Linear(self.hidden_size *2, self.hidden_size)
        self.dropout=nn.Dropout(self.dropout_p)
        self.gru=nn.GRU(self.hidden_size, self.hidden_size)
        self.out=nn.Linear(self.hidden_size, self.output_size)
    def forward(self, input, hidden,encoder_outputs):
        embedded=self.embedding(input).view(1,1,-1) # 임베딩
        embedded=self.dropout(embedded)
        # 언텐션 가중치(그 전 은닉상태와 입력을 통해)
        attn_weights = f.softmax(self.attn(torch.cat((embedded[0], hidden[0]), 1)), dim=1)
        attn_applied=torch.bmm(attn_weights.unsqueeze(0),
                              encoder_outputs.unsqueeze(0))
        # 인코더 결과와 어넽션 가중치를 통해서 bmm함수 사용)
        output=torch.cat((embedded[0],attn_applied[0]),1)
        output=self.attn_combine(output).unsqueeze(0)
        output=f.relu(output)
        output,hidden=self.gru(output, hidden)
        output=f.log_softmax(self.out(output[0]),dim=1)
        return output, hidden, attn_weights
    def initHidden(self):
        return torch.zeros(1,1, self.hidden_size, device=device)
def preprocess(df):
    pairs=[]
    source_vocab=Vocab()
    target_vocab=Vocab()
    for i in df.index:
        source_vocab.add_vocab(df['Q'][i])
        target_vocab.add_vocab(df['A'][i])
        temp=[df['Q'][i],df['A'][i]]
        pairs.append(temp)
    
    return source_vocab, target_vocab,pairs

source_vocab, target_vocab, pairs=preprocess(df)


enc_hidden_size=10
dec_hidden_size=enc_hidden_size

enc=EncoderA(source_vocab.n_vocab,enc_hidden_size)
dec=AttnDecoder(dec_hidden_size,target_vocab.n_vocab)

def tensorize(vocab,sentence):
    idx=[vocab.vocab2index[word] for word in sentence.split(' ')]
    idx.append(vocab.vocab2index['<EOS>'])
    return torch.Tensor(idx).long().to(device).view(-1,1)


train_source=[tensorize(source_vocab,pair[0]) for pair in pairs]
train_target=[tensorize(target_vocab,pair[1]) for pair in pairs]
loss_total=0
number_epoch=101
# 학습시키기. 
encoder_optimizer=optim.SGD(enc.parameters(),lr=0.01)
decoder_optimizer=optim.SGD(dec.parameters(),lr=0.01)
criterion=nn.NLLLoss()
for epoch in range(number_epoch):
    epoch_loss=0
    for i in range(len(train_source)):
        encoder_optimizer.zero_grad()
        decoder_optimizer.zero_grad()
        source_tensor=train_source[i]
        target_tensor=train_target[i]
        encoder_hidden=torch.zeros([1,1,enc.hidden_size]).to(device)
        source_length=source_tensor.size(0)
        target_length=target_tensor.size(0)
        
        encoder_outputs=torch.zeros(MAX_LENGTH, enc.hidden_size, device=device)
        loss=0
        
        for ecount in range(source_length):
            encoder_output, encoder_hidden = enc(source_tensor[ecount], encoder_hidden)
            
        decoder_input = torch.Tensor([[SOS_token]]).long().to(device)
        decoder_hidden = encoder_hidden 
        for dcount in range(target_length):
            decoder_output, decoder_hidden,decoder_attention = dec(decoder_input, decoder_hidden,encoder_outputs)       
            loss += criterion(decoder_output, target_tensor[dcount])
            decoder_input = target_tensor[dcount]  
        
        loss.backward()
        encoder_optimizer.step()
        decoder_optimizer.step()
        epoch_loss += loss.item()/target_length
    if epoch % 100 == 0:
        print('--- epoch {}, total loss {} '.format(epoch,float(epoch_loss/15))) 
        
        
# 성능 평가하기. 
for pair in pairs:
    source_tensor=tensorize(source_vocab, pair[0])
    source_length=source_tensor.size()[0]
    encoder_hidden=torch.zeros([1,1,enc.hidden_size]).to(device)
    for ecount in range(source_length):
        _,encoder_hidden=enc(source_tensor[ecount],encoder_hidden)
    decoder_input = torch.Tensor([[SOS_token]], device=device).long()
    decoder_hidden = encoder_hidden
    decoded_words = []
    decoder_attentions=torch.zeros(MAX_LENGTH,MAX_LENGTH)
    for dcount in range(20):
        decoder_output,decoder_hidden,decoder_attention=dec(decoder_input,decoder_hidden,encoder_outputs)
        decoder_attentions[dcount]=decoder_attention.data
        _, top_index=decoder_output.data.topk(1)
        if top_index.item()==EOS_token:
            decoded_words.append('<EOS>')
            break
        else:
            decoded_words.append(target_vocab.index2vocab[top_index.item()])
            
        decoder_input=top_index.squeeze().detach()
    predict_words=decoded_words
    predict_sentence=" ".join(predict_words)
    print("Q",pair[0])
    print("A",pair[1])
    print("Predict words",predict_sentence)    
 
```
